```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

Problem 0

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

Problem 1

```python
x = np.linspace(0,len(bkgd),1000)
y = bkgd[0:1000]
plt.scatter(x,y)
```

```python
def gen_scat(start,end):
    x = np.linspace(start,end,end-start)
    y = bkgd[start:end]
    plt.scatter(x,y)
```

```python
gen_scat(500,2000)
```

```python
def gen_scat_100(start, end):
    x = np.linspace(start,end,end-start)
    y = bkgd[0::100]
    plt.scatter(x,y)
```

```python
gen_scat_100(0,1000)
```

```python
x = np.linspace(0,len(bkgd),1000)
y = bkgd[0:1000]
dist = plt.hist2d(x,y,bins=10)
plt.show()
```

B. Each color represents the density. This is a 2d color plot which means the changing colors represent another dimension. In this case the yellow is the most dense areas and the dark blue/violet are the less dense areas.
